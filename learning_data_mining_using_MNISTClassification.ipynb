**Non-negative Matrix Factorization or NMF**

```python

from keras.datasets import mnist
from sklearn.decomposition import NMF
import numpy as np



X_train_normalized = X_train / 255.0
X_test_normalized = X_test / 255.0


X_train_flat = X_train_normalized.reshape(X_train_normalized.shape[0], -1)
X_test_flat = X_test_normalized.reshape(X_test_normalized.shape[0], -1)

# Apply NMF
# Initialize NMF
nmf = NMF(n_components=50)  # Number of components to extract

# Fit NMF to the training data
W_train = nmf.fit_transform(X_train_flat)

# Transform the test data using the fitted NMF model
W_test = nmf.transform(X_test_flat)

# W_train and W_test now contain the NMF features for training and testing

# Now W_train contains the weights to reconstruct the training images from the NMF components,
# and H contains the basis components (features).
```

**Visualization**

```python
import matplotlib.pyplot as plt

# 'nmf' is our fitted NMF model
H = nmf.components_

# Set the number of components you want to display
n_display = 5

fig, axes = plt.subplots(1, n_display, figsize=(20, 4))
for i, ax in enumerate(axes.flat):
    if i < 5:  # Only display the first 5 components
        ax.imshow(H[i].reshape(28, 28), cmap='gray')
        ax.axis('off')
plt.tight_layout()
plt.show()
```

**Note**
We now plot the components (the ( H ) matrix) to see what kind of features NMF has identified. Each component can be reshaped back into the original image size and displayed.

**Reconstructing images**

I try to reconstruct images from the NMF features to see how well the NMF components capture the essence of the data.

```python
# Choose an image to reconstruct
image_index = 0

# Reconstruct the image from the NMF components
reconstructed_image = np.dot(W_train[image_index], H)

# Display the original and reconstructed images
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(X_train_flat[image_index].reshape(28, 28), cmap='gray')
plt.title('Original Image')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(reconstructed_image.reshape(28, 28), cmap='gray')
plt.title('Reconstructed Image')
plt.axis('off')

plt.tight_layout()
plt.show()
```

**Using Random Forest Classifier**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report

# Train a classifier
clf = RandomForestClassifier()
clf.fit(W_train, y_train)

# Predict on the test set
y_pred = clf.predict(W_test)

# Evaluate the results
print(classification_report(y_test, y_pred))
```

```python
from tensorflow.keras import models, layers

cnn_model = models.Sequential([
# First, we define the input shape in the first layer only
layers.Conv2D(filters=16, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)),
layers.MaxPooling2D(pool_size=(2,2)),

# Adding another set of convolution and max pooling layers
layers.Conv2D(filters=32, kernel_size=(3,3), activation='relu'),
layers.MaxPooling2D(pool_size=(2,2)),

# Flatten the output of the conv layers to feed into the dense layers
layers.Flatten(),

# Dense layers
layers.Dense(50, activation='relu'),
layers.Dense(10, activation='softmax')  # Use 'softmax' for multi-class classification
])

# Compile the CNN model
cnn_model.compile(
optimizer='adam',
loss='sparse_categorical_crossentropy',
metrics=['accuracy']
)

# Fit the model on the training data
history = cnn_model.fit(X_train_normalized, y_train, epochs=10,validation_data=(X_test_normalized, y_test))
```

**CONVOLUTIONAL AUTOENCODER**

```python
from tensorflow.keras.layers import Input, Dense, Conv2D, MaxPooling2D, UpSampling2D
from tensorflow.keras.models import Model

# Define the input shape
input_img = Input(shape=(28, 28, 1))

# Encoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(input_img)
x = MaxPooling2D((2, 2), padding='same')(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
encoded = MaxPooling2D((2, 2), padding='same')(x)

# Decoder
x = Conv2D(32, (3, 3), activation='relu', padding='same')(encoded)
x = UpSampling2D((2, 2))(x)
x = Conv2D(32, (3, 3), activation='relu', padding='same')(x)
x = UpSampling2D((2, 2))(x)
decoded = Conv2D(1, (3, 3), activation='sigmoid', padding='same')(x)

# Autoencoder model
autoencoder = Model(input_img, decoded)
autoencoder.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Train the autoencoder model and save the history
history = autoencoder.fit(X_train_normalized, X_train_normalized,
epochs=10,  # Adjust the number of epochs if needed
batch_size=64,
shuffle=True,
validation_data=(X_test_normalized, X_test_normalized))
```

```python
from tensorflow.keras.layers import Input, Dense, Flatten
from tensorflow.keras.models import Model

# 'autoencoder' the trained model and 'X_train_normalized' and 'X_test_normalized' are data

# Extract Encoded Features
# Get the encoder part of the autoencoder
# The name of the encoded layer is not 'encoded', so we reference it by its index
encoder_output = autoencoder.layers[-7].output
encoder = Model(inputs=autoencoder.input, outputs=encoder_output)

# Encode the training and test data
X_train_encoded = encoder.predict(X_train_normalized)
X_test_encoded = encoder.predict(X_test_normalized)

# Flatten the encoded images for the classifier
X_train_encoded_flat = X_train_encoded.reshape(X_train_encoded.shape[0], -1)
X_test_encoded_flat = X_test_encoded.reshape(X_test_encoded.shape[0], -1)

# Train a Classifier
# Define the classifier model
classifier_input = Input(shape=(X_train_encoded_flat.shape[1],))
classifier_output = Dense(10, activation='softmax')(classifier_input)
classifier_model = Model(classifier_input, classifier_output)

# Compile the model
classifier_model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])

# Train the classifier
classifier_model.fit(X_train_encoded_flat, y_train, epochs=10,  # Adjust the number of epochs if needed
batch_size=64,
validation_data=(X_test_encoded_flat, y_test))

# Evaluate the Classifier
# Evaluate the model on the test set
test_loss, test_accuracy = classifier_model.evaluate(X_test_encoded_flat, y_test)

print(f"Test accuracy: {test_accuracy}")
```
